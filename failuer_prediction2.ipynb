## 問題の前提確認

$t\in R^{＋}$に対して確率密度関数$f_{T}(t)$を持つ非負の連続型確率変数$T$を考える。$T = T_{1},...,T_{n}$は確率密度関数$f_{T}(t)$持つ分布からの独立同一な大きさ$n$の標本である。また、$t= t_{1},...,t_{n}$をその実現値とする。$i$番目の個体が故障の時$\delta_{i}=1$,右打ち切りの時$\delta_{i}=0$となる2値変数を考える。

観測された時間データに対して$T～WEIB(\mu,\sigma)$を仮定する。この時$-\infty<\mu<\infty,0<\sigma<\infty$である。この時の累積分布関数と確率密度関数は以下の用に示せる。

\begin{align}
F_{T}(t;{\mu},{\sigma}) = \Phi_{sev}{\bigg (}{\frac {\log{t}-{\mu}}{\sigma}}{\bigg)}\end{align}

\begin{align}f_{T}(t;{\mu},{\sigma}) = {\frac {1}{t{\sigma}}}\Phi_{sev}{\bigg (}{\frac {\log{t}-{\mu}}{\sigma}}{\bigg)},　t>0\end{align}



ここで
\begin{align}
\Phi_{sev}(z)=1-e^{-e^{z}}\end{align}
\begin{align}\phi_{sev}(z)= e^{z-e^{z}}　,　-\infty<z<\infty \end{align}



このような前提条件のもと問題を解いていく。

# 解答結果

## **1.(a)**
このモデルの尤度関数を求める。
\begin{align} L = \prod_{i=1}^{n}　f_{t_{i}}^{\delta_{i}}{(}1-F_{t}{)}^{1-\delta_{i}}\end{align}

この尤度化数の対数尤度関数を求める。

\begin{align}
 \log{L} &= \sum_{i=1}^{n}{\delta_{i}}log{f_{t_{i}}}+\sum_{i=1}^{n}({1-\delta_{i}})log{(}1-F_{t}{)}\\
 & = \sum_{i=1}^{n}\delta_{i}{(}\log{e^{z-e^z}}-\log{t_{i}\sigma}{)}
 +\sum_{i=1}^{n}{(}1-\delta_{i}{)}\log{e^{-e^z}}\\
 & = \sum_{i=1}^{n}\delta_{i}\{(z-e^z)-\log{t_{i}\sigma}\}+\sum_{i=1}^{n}(\delta_{i}-1)e^z\\
 & =\sum_{i=1}^{n}(\delta_{i}z-\delta_{i}\log{t_{i}\sigma})-\sum_{i=1}^{n}e^z
\end{align}



対数尤度関数の偏微分を行う。
 \begin{align}
 \log{L}^{(1)}(t) &=-\frac{1}{\sigma}\sum_{i=1}^{n}\delta_{i}+\frac{1}{\sigma}\sum_{i=1}^{n}e^{\frac{log{t_{i}}-\mu}{\sigma}}\end{align}
\begin{align}
 \log{L}^{(1)}(t) &=-\frac{1}{\sigma}\sum_{i=1}^{n}\delta_{i}+\frac{1}{\sigma}\sum_{i=1}^{n}e^{\frac{log{t_{i}}-\mu}{\sigma}}=0\end{align}
 を$\mu$について解く。

 \begin{align}
 \sum_{i=1}^{n}\delta_{i}&=e^{\frac{\sum_{i=1}^{n}log{t_{i}}-\mu}{\sigma}}\\
 \log{\sum_{i=1}^{n}\delta_{i}}&=\frac{\sum_{i=1}^{n}log{t_{i}}-\mu}{\sigma}\\
 \frac{\mu}{\sigma}& = \frac{\log{\sum_{i=1}^{n}}t_{i}^{\frac{1}{\sigma}}}{\sum_{i=1}^{n}\log{\delta_{i}}}\\
 \hat\mu& =  \sigma\frac{\log{\sum_{i=1}^{n}}t_{i}^{\frac{1}{\sigma}}}{\sum_{i=1}^{n}\log{\delta_{i}}}\end{align}

## **1(b)**
対数尤度関数を二回偏微分を行う。

\begin{align}
 \log{L}^{(1)}(t) &=-\frac{1}{\sigma}\sum_{i=1}^{n}\delta_{i}+\frac{1}{\sigma}\sum_{i=1}^{n}e^{\frac{log{t_{i}}-\mu}{\sigma}}\end{align}
 これより
 \begin{align}
 \log{L}^{(2)}(t) &=-\frac{1}{\sigma^2}\sum_{i=1}^{n}t_{i}^{\frac{1}{\sigma}}e^{\frac{-\mu}{\sigma}}\\
 & = -\frac{1}{\sigma^2}\sum_{i=1}^{n}t_{i}^{\frac{1}{\sigma}}e^{\frac{- \sigma\frac{\log{\sum_{i=1}^{n}}t_{i}^{\frac{1}{\sigma}}}{\sum_{i=1}^{n}\log{\delta_{i}}}}{\sigma}}\\
 & =-\frac{1}{\sigma^2}\sum_{i=1}^{n}t_{i}^{\frac{1}{\sigma}}\frac{\sum_{i=1}^{n}\delta_{i}}{\sum_{i=1}^{n}t_{i}^{\frac{1}{\sigma}}}\\
 & =- \frac{\sum_{i=1}^{n}\delta_{i}}{\sigma^2}\end{align}



以上より二回偏微分のマイナス（フィッシャー情報量）の逆数が漸近分散になるので
\begin{align}\widehat{aVar{(}\hat{\mu}{)}}= \frac{\sigma^2}{\sum_{i=1}^{n}\delta_{i}}\end{align}

## 数値計算による確認

ワイブル分布を仮定しているため、最小極値分布を定義する。そして、ワイブル分布は最小極値分布の対数ロケーションスケールファミリーであることを利用分布関数と密度関数を定義する。

そして、上記に示した対数尤度関数を定義し、$\mu$をパラメータとして考える。
定義した対数尤度関数をネルダーミード法によって最大化する。（ネルダーミード法は初期値に強い）

```r
psev = function(z){ 1-exp(-exp(z))} # 最小極値分布の累積分布関数
dsev = function(z){ exp(z-exp(z)) } # 最小極値分布の確率密度関数
Ft = function(t, mu){ psev((log(t)-mu)/sigma) } # ワイブル分布の累積分布関数
ft = function(t, mu){ (1/(t*sigma)) * dsev((log(t)-mu)/sigma) } # ワイブル分布の確率密度関数
l_i = function(x, par, n_d, n_r){ n_d * log(ft(x, par)) + n_r * log(1-Ft(x, par)) } # 1点の対数尤度.分布はワイブル分布を仮定
loglikelihood = function(x, n_d, n_r){# 対数尤度
function(par){ sum(l_i(x, par, n_d, n_r)) }
}
l = loglikelihood(d$t, d$dj, 1-d$dj) # 対数尤度の定義
res = optim(20, l, control = list(fnscale=-1), hessian = TRUE) # optim 関数で対数尤度を最大化
hat.mu = res$par # 最尤推定値
hat.mu

H_n = res$hessian # ヘッセ行列
hat.fisher.information = -H_n # 観測フィッシャー情報行列
hat.aVar = solve(hat.fisher.information) # プラグイン漸近分散
hat.aVar
```

### 累積分布関数の95%近似信頼区間

$I_{n}^{-1}=\widehat{aVar(\hat\mu)}$である。


$F_{t}(t;\hat{\mu})～N(F_{t}(t;\hat{\mu}),J_{F_{t}}(\hat{\mu})I_{n}^{-1}(\hat\mu)J_{F_{t}}^\top)$

95％信頼区間は$F_{t}\pm  z_{\alpha}\sqrt{J_{F_{t}}(\hat{\mu})I_{n}^{-1}(\hat\mu)J_{F_{t}}^\top}$と表すことが出来る。

ここで必要なヤコビ行列を求める。
\begin{align}\frac{\partial F_{t}(t;\mu)}{\partial \mu}&= \frac{\partial}{\partial \mu}(1-e^{-e^{\frac{\log{t}-\mu}{\sigma}}})\\
& = e^{-e^{\frac{\log{t}-\mu}{\sigma}}}\frac{\partial}{\partial \mu}e^{\frac{\log{t}-\mu}{\sigma}}\\
& = -\frac{1}{\sigma}e^{\frac{\log{t}-\mu}{\sigma}-e^{\frac{\log{t}-\mu}{\sigma}}}\end{align}

```r
hat.F = function(t) {psev((log(t)-hat.mu)/sigma)}
var.F = function(t){
  (1/sigma)^2*exp(2*((log(t)-hat.mu)/sigma - exp((log(t) - hat.mu)/sigma)))*aVar
}
t = seq(1, 30000, len=100)
dFT = data.frame(t,
mean=hat.F(t),
upper=hat.F(t)+qnorm(0.975)*sqrt(var.F(t)),
lower=hat.F(t)-qnorm(0.975)*sqrt(var.F(t)))
#dFT = dFT %>% mutate(upper = ifelse(upper>1, 1, upper))
plot = dFT %>%
ggplot()+geom_line(aes(t, mean))+
geom_line(aes(t, upper), linetype=2)+
geom_line(aes(t, lower), linetype=2)+
xlab(bquote(italic(t)))+ylab("Cumulative␣Distribution␣Function")
plot
```

```r
ci.F = function(t){
  data.frame(t,
            mean=hat.F(t),
            upper=hat.F(t)+qnorm(0.975)*sqrt(var.F(t)),
            lower=hat.F(t)-qnorm(0.975)*sqrt(var.F(t)))
}
ci.F(10000)
```

これより累積分布関数の95%近似信頼区間がわかった。

$0.3360<F_{10000}(t;\hat\mu,\sigma)<0.72153$

## **1(f)**

### 10%点の95％近似信頼区間

上側パーセント点は次のように表せる。$t_{p}=\exp(\hat{\mu}+\sigma\log{(-\log(1-p))})$
そして、$t_{p}$は$t_{p}(\hat{\mu})～N(t_{p}(\hat{\mu}),J_{t_{p}}(\hat{\mu})I_{n}^{-1}(\hat\mu)J_{t_{p}}^\top)$に従う。

ここでヤコビ行列を求める
\begin{align}\frac{\partial t_{p}(\mu)}{\partial \mu}&= \frac{\partial}{\partial \mu}\exp(\mu+\sigma\log{(-\log(1-p))})\\
& = \exp(\mu+\sigma\log{(-\log(1-p))})\\
& = t_{p}\end{align}

以上より
\begin{align}\widehat{aVar(t_{p}(\mu))} = t_{p}^2\times \widehat{aVar(\hat{\mu})}\end{align}



つまり、95％信頼区間は$t_{p}\pm  z_{\alpha}\sqrt{J_{t_{p}}(\hat{\mu})I_{n}^{-1}(\hat\mu)J_{t_{p}}^\top}$と表すことが出来る。

```r
hat.t = function(p) {exp(hat.mu+sigma*log(-log(1-p)))}
var.t = function(p){
  exp(hat.mu+sigma*log(-log(1-p)))^2*aVar
}
p = seq(0, 1,len = 1000)
dFT = data.frame(t,
mean=hat.t(p),
upper=hat.t(p)+qnorm(0.975)*sqrt(var.t(p)),
lower=hat.t(p)-qnorm(0.975)*sqrt(var.t(p)))
#dFT = dFT %>% mutate(upper = ifelse(upper>1, 1, upper))
plot = dFT %>%
ggplot()+geom_line(aes(p, mean))+
geom_line(aes(p, upper), linetype=2)+
geom_line(aes(p, lower), linetype=2)+
xlab(bquote(italic(p)))+ylab("Cumulative␣Distribution␣Function")
plot
```

```r
ci.t = function(p){
  data.frame(p,
            mean=hat.t(p),
            upper=hat.t(p)+qnorm(0.975)*sqrt(var.t(p)),
            lower=hat.t(p)-qnorm(0.975)*sqrt(var.t(p)))
}
ci.t(0.1)
```

以上の結果より$t_{p}$の95％信頼区間は$2724.877<t_{0.1}<4758.973$であることがわかった。

### **$\mu$と$\sigma$の数値計算による最尤推定と信頼区間を求める**

問題１の時定義した対数尤度関数を用いる。さらに、今回は問題１の数値計算の時では$\mu$だけが未知であったが、今回は$\mu$,$\sigma$ともに未知であるため、二つをパラメータとして設定する。そして、同様にoptim関数を用いて対数尤度関数を最大化するパラメータを求める。

```r
li_for_llsf = function(t, par, delta_i){
mu = par[1]
sigma = par[2]
pdf = 1/(sigma*t) * phi((log(t)-mu)/sigma);
cdf = Phi((log(t)-mu)/sigma)
delta_i * log(pdf) + (1-delta_i) * log(1-cdf)
}
loglikelihood = function(t, delta){# 対数尤度
  function(par){
    sum(li_for_llsf(t, par, delta))
  }
}
```

```r
psev = function(z){ 1-exp(-exp(z)) }
dsev = function(z){ exp(z-exp(z)) }
qsev = function(p){ log(-log(1-p)) }


Phi = psev # ワイブル分布
phi = dsev
l = loglikelihood(d$t, d$dj)
l
l(c(10, 1))
```

```r
### 最尤推定
initial_par = c(10, 1)
res = optim(initial_par, l, control = list(fnscale=-1), hessian = TRUE)
res
### 最尤推定値
hat.mu = res$par[1]; hat.mu
hat.sigma = res$par[2]; hat.sigma ### プラグイン漸近分散
H_n = res$hessian
fisher.information = -H_n
aVar = solve(fisher.information); sqrt(aVar) ### 95%近似信頼区間
c(hat.mu-qnorm(0.975) * sqrt(aVar[1,1]),
  hat.mu+qnorm(0.975) * sqrt(aVar[1,1]))
c(hat.sigma-qnorm(0.975) * sqrt(aVar[2,2]),
hat.sigma+qnorm(0.975) * sqrt(aVar[2,2]))
```

## 2f

### 10%点の95％近似信頼区間

ロケーションスケールファミリーの上側pパーセント点
\begin{align}t_{p} =t_{p}(\mu,\sigma)=\mu + \sigma\Phi^{-1} _{p}\end{align}
で表せる。今回はワイブル分布であるため対数ロケーションファミリーとして表すことが出来る。
\begin{align}t_{p} =t_{p}(\mu,\sigma)=\exp(\mu + \sigma\Phi^{-1} _{p})\end{align}


そして、パラメータが二つあることに注意してヤコビ行列を求める。

\begin{align}\frac{\partial t_{p}(\theta)}{\partial \theta}&=
\bigg{(}t_{p}(\mu,\sigma)\frac{\partial}{\partial \mu}[\mu +\sigma\Phi^{-1}_{p}])
t_{p}(\mu,\sigma)\frac{\partial}{\partial \sigma}[\mu +\sigma\Phi^{-1}_{p}]\bigg{)}\\
& = \bigg{(}t_{p}(\mu,\sigma)　\Phi^{-1}_{p}t_{p}(\mu,\sigma)\bigg{)}\end{align}

以上よりプラグイン漸近分散は
\begin{align}\widehat{aVar(t_{p}(\mu,\sigma))} = J_{t_{p}}(\hat{\mu},\hat{\sigma})I_{n}^{-1}(\hat\mu,\hat{\sigma})J_{t_{p}}^\top
\end{align}


```r
inv.Phi = qsev
hat.tp = function(p){
  exp(hat.mu + hat.sigma*inv.Phi(p))
}
var.tp = function(p){
  jtp_mu = hat.tp(p)
  jtp_sigma = inv.Phi(p)*hat.tp(p)
  jtp = matrix(c(jtp_mu,jtp_sigma),ncol = 2)
  jtp%*%aVar%*%t(jtp)
}
```

```r
p = seq(0, 1, len=100)
#sqrt(var.tp(p))
dtp = data.frame(p,
mean = hat.tp(p),
lower= sapply(p, function(p) hat.tp(p)-qnorm(0.975)*sqrt(var.tp(p))),
upper= sapply(p, function(p) hat.tp(p)+qnorm(0.975)*sqrt(var.tp(p)))
)
p1 = dtp %>%
ggplot()+geom_line(aes(p, mean))+
geom_line(aes(p, upper), linetype=2)+
geom_line(aes(p, lower), linetype=2)+
xlab(bquote(italic(p)))+ylab("quantile")
p1
```

```r
ci.t = function(p){
  data.frame(p,
            mean=hat.tp(p),
            upper=hat.tp(p)+qnorm(0.975)*sqrt(var.tp(p)),
            lower=hat.tp(p)-qnorm(0.975)*sqrt(var.tp(p)))
}
ci.t(0.1)
```

以上の結果より$t_{p}$の95％信頼区間は\begin{align}2194.819<t_{0.1}<4319.308&\end{align}であることがわかった。
