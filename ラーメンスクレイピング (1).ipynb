```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

url="https://ramendb.supleks.jp/rank?state=tokyo"
res=requests.get(url)
soup=BeautifulSoup(res.content,"html.parser")
spot=soup.find("div",attrs={"class":"ranks"})

spot1=spot.find("table")
trs = spot1.find_all('tr')

url_list=[]
for tr in trs :
  a_tag=tr.find("a")
  url_list.append(a_tag)

link_list=[]
for link in url_list:
 link_list.append(link.get('href'))

for url in link_list:
  url_new=("https://ramendb.supleks.jp/"+url)


adress=[]
name=[]
for url in link_list:
  url_new=("https://ramendb.supleks.jp/"+url)
  res=requests.get(url_new)
  soup=BeautifulSoup(res.content,"html.parser")
  data=soup.find("div",attrs={"class":"datas"})
  table=data.find("table")
  trs=table.find_all("tr")
  name.append(trs[0].find("td"))
  adress.append(trs[1].find("td"))

text_adress=[]
for ad in adress:
  text_adress.append(ad.find("span").get_text())

import re
correct_text=[]
for text in text_adress:
  text_mod = re.sub('^〒([0-9]{3})-([0-9]{4})','', text)
  correct_text.append(text_mod)


text_name=[]
for na in name:
  text_name.append(na.find("span").get_text())

df=pd.DataFrame(data=text_name)
df[1]=correct_text
df.index = np.arange(1, len(df)+1)


with open("correct_ra-men_data.csv", mode="w", encoding="utf-8", errors="ignore") as f:
        df.to_csv(f)
```
